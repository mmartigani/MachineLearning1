```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,2,100)
x
```

```python
fig,ax=plt.subplots()
ax.plot(x,x,label="linear")
ax.plot(x,x**2,label="cuadratica")
ax.plot(x,x**3,label="cubica")
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")
ax.set_title("Grafico")
ax.legend()
```

```python
x=np.linspace(0,2,100)
plt.plot(x,x,label="linear")
plt.plot(x,x**2,label="cuadratica")
plt.plot(x,x**3,label="cubica")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Grafico")
plt.legend()
```

```python
def plot_function(ax, data1, data2, param_dict):
  #funcion para hacer grafico
  #ax = ejes de dibujo
  #data1 = array vector de datos
  #data2 = array vector de datos
  #param_dict=dicccionario de parametros
  out=ax.plot(data1,data2,**param_dict)
  return out
```

```python
fig,ax=plt.subplots(1,1)
data_a =[1,2,3,4]
data_b = [2,3,4,5]
par_dict = {"marker":"o", "color":"green", "markersize":12}
plot_function(ax=ax, data1 =data_a, data2=data_b,param_dict=par_dict)
```

```python
fig,ax=plt.subplots(1,1)
data_c =[10,20,30,40]
data_d = [100,200,300,400]
par_dict = {"marker":"o", "color":"green", "markersize":12}
plot_function(ax=ax, data1 =data_c, data2=data_d,param_dict=par_dict)
```

```python
fig,(ax1, ax2)=plt.subplots(1,2)
plot_function(ax=ax1, data1 =data_a, data2=data_b,param_dict= {"marker":"x", "color":"red", "markersize":12})
plot_function(ax=ax2, data1 =data_c, data2=data_d,param_dict= {"marker":"o", "color":"green", "markersize":12})
```
