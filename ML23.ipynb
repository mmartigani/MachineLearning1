```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
inertias = []
n_clusters = range(1,10)
for n in n_clusters:
  print("evaluando n_clusters = ", n)
  knn_model = KMeans(n_clusters=n)
  knn_model.fit(df)
  inertias.append(knn_model.inertia_)
```

```python
inertias
```

```python
plt.figure(figsize=(16,8))
plt.plot(n_clusters, inertias)
plt.xlabel("nro de clusters")
plt.ylabel("inercias")
plt.show()
```

```python
iris.target
```

```python
knn_final= KMeans(n_clusters=3)
knn_final.fit(df)
```

```python
df["clusters_id"] = knn_final.predict(df)
df
```

```python
df["iris_class"]=iris.target
df
```

```python
fig,axes = plt.subplots(1,2, figsize=(16,8))
axes[0].scatter(df["sepal length (cm)"],df["sepal width (cm)"],c=df.iris_class)
axes[0].set_title("iris_class")
axes[1].scatter(df["sepal length (cm)"],df["sepal width (cm)"],c=df.clusters_id)
axes[1].set_title("clusters id")
```
