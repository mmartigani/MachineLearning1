```python
uno = pd.DataFrame({"col1":[1,2,3],
                    "col2":[4,5,6],
                    "col3":[7,8,9]},
                   index=["id1","id2","id3"])
uno
```

```python
dos= pd.DataFrame({"col1":[10,11,12,13],
                    "col2":[14,15,16,17],
                    "col3":[18,19,20,21]})

dos
```

```python
tres = pd.DataFrame({"col1":["a","b","c"],
                    "col2":["d","e","f"],
                    "col3":["g","h","i"],
                     "col4":["j","k","l"]},
                   index=["n1","n2","n3"])
tres
```

```python
result1 = pd.concat([uno,dos],
                    axis=0,
                    join="outer",
                    ignore_index=False)
result1
```

```python
result2 = pd.concat([uno,dos],
                    axis=0,
                    join="outer",
                    ignore_index=True)
result2
```

```python
result3 = pd.concat([uno,tres],
                    axis=0,
                    join="outer",
                    ignore_index=False)
result3
```

```python
result4 = pd.concat([uno,tres],
                    axis=0,
                    join="inner",
                    ignore_index=False)
result4
```

```python
result5 = pd.concat([uno,tres],
                    axis=1,
                    join="outer",
                    ignore_index=False)
result5
```

```python
serie1=pd.Series(["s1","s2","s3"],
                 name="col5",
                 index=["n1", "n2","n3"])
serie1
```

```python
result6=pd.concat([tres, serie1],
                  axis=1,
                  join="outer",
                  ignore_index=False)
result6
```

```python
tres.merge(cuatro, how='inner', on="col1")
```
