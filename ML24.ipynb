```python
#Decision Tree (arboles de decision)
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn import tree
```

```python
iris = datasets.load_iris()
```

```python
iris.target
```

```python
estimator=DecisionTreeClassifier()
```

```python
param_grid ={"min_samples_split":range(2,11,1),
             "min_samples_leaf":[1,3,5]}
```

```python
gs =GridSearchCV(estimator=estimator,
                 param_grid=param_grid,
                 scoring=None,
                 refit=True,
                 cv=None)
```

```python
gs.fit(iris.data, iris.target)
```

```python
pd.DataFrame(gs.cv_results_)
```

```python
gs.best_params_
```

```python
new_data =np.array([[5.3,3.5,1.2,0.5],
                    [4.8,3.1,1.3,0.4]])
```

```python
iris.target
```

```python
gs.best_estimator_.predict(new_data)
```

```python
gs.best_estimator_.predict_proba(new_data)
```

```python
gs.best_estimator_.get_depth()
```

```python
gs.best_estimator_.feature_importances_
```

```python
tree.plot_tree(gs.best_estimator_)
```

```python
# Regesion + Grid Search cv
from sklearn import datasets
import graphviz
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.model_selection import GridSearchCV, train_test_split
```

```python
diabetes = datasets.load_diabetes()
```

```python
diabetes.data
```

```python
diabetes.target
```

```python
regressor = tree.DecisionTreeRegressor()
```

```python
param_grid={"max_depth":range(1,101)}
```

```python
gs=GridSearchCV(estimator=regressor,
                param_grid=param_grid,
                scoring=None,
                cv=None)
```

```python
X_train, X_test, y_train, y_test= train_test_split(diabetes.data,diabetes.target,test_size=0.1 )
```

```python
gs.fit(X_train,y_train)
```

```python
pd.DataFrame(gs.cv_results_)
```

```python
best_model=gs.best_estimator_
best_model
```

```python
best_model.predict(X_test)
```

```python
pd.DataFrame({"y_test":y_test,"y_pred":best_model.predict(X_test)})
```

```python
best_model.score(X_test,y_test)
```

```python
data1=tree.export_graphviz(best_model,out_file=None)
```

```python
graph1=graphviz.Source(data1)
```

```python
graph1.render("diabetes_best")
```
